```python
# import libraries 
import nltk
```

```python
# loading the datset
names =nltk.corpus.names
```

```python
# to check the files in the dataset
names.fileids()
```

```python
# Visualizing 
cfd = nltk.ConditionalFreqDist(
(fileid, name[-1])
for fileid in names.fileids()
for name in names.words(fileid))
```

```python
cfd.plot()
```

```python
# function to return last letter of the name 
def gender_feature(word):
    return{'lastletter': word[-1]}
```

```python
gender_feature('Shrek')
```

```python
# making a list of name and target variable
import random
names = ([(name, 'male') for name in names.words('male.txt')] +
[(name, 'female') for name in names.words('female.txt')])
random.shuffle(names)
```

```python
names
```

```python
# selecting the features ( last letter of an alphabet)
featuresets = [(gender_feature(n), g) for (n,g) in names]
```

```python
featuresets
```

```python
# splitting the data 
train_set, test_set = featuresets[500:], featuresets[:500]
```

```python
# model 
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
 # classifying 
classifier.classify(gender_feature('Neo'))
```

```python
classifier.classify(gender_feature('Trinity'))
```
